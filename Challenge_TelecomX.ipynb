**Distribuci√≥n de evasi√≥n**

comprender c√≥mo est√° distribuida la variable "churn" (evasi√≥n) entre los clientes. Utiliza gr√°ficos para visualizar la proporci√≥n de clientes que permanecieron y los que se dieron de baja.

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Contar valores 0 y 1
churn_counts = df_normalizado['abandono'].value_counts()

# Mapear los valores 0 y 1 a nombres legibles
label_map = {0: 'Permanecen', 1: 'Dieron de Baja'}
labels = [label_map.get(i, str(i)) for i in churn_counts.index]

# Gr√°fico de pastel
plt.figure(figsize=(6, 6))
plt.pie(
    churn_counts,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#66c2a5', '#fc8d62']
)
plt.title('Proporci√≥n de Clientes: Permanecen vs Dieron de Baja')
plt.axis('equal')
plt.show()
```

**Recuento de evasi√≥n por variables categ√≥ricas**

Exploraremos c√≥mo se distribuye la evasi√≥n seg√∫n variables categ√≥ricas, como g√©nero, tipo de contrato, m√©todo de pago, entre otras.

Este an√°lisis puede revelar patrones interesantes, por ejemplo, si los clientes de ciertos perfiles tienen una mayor tendencia a cancelar el servicio, lo que ayudar√° a orientar acciones estrat√©gicas.

```python
sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df_normalizado, x='genero', hue='abandono', palette='Set2')

plt.title("Porcentaje de Evasi√≥n por G√©nero", fontsize=16)
plt.xlabel("G√©nero", fontsize=12)
plt.ylabel("Cantidad de Clientes", fontsize=12)
plt.legend(title='Abandono', fontsize=10, title_fontsize=12)

# Crear tabla resumen para valores y porcentajes
tabla = df_normalizado.groupby(['genero', 'abandono']).size().reset_index(name='count')
total_por_genero = df_normalizado.groupby('genero').size().reset_index(name='total')

# Diccionario para acceder total por g√©nero
total_dict = dict(zip(total_por_genero['genero'], total_por_genero['total']))

# Los grupos en el eje x (g√©neros)
generos = df_normalizado['genero'].unique()
# Los niveles de hue ('S√≠', 'No')
abandono_niveles = df_normalizado['abandono'].unique()

# Para cada barra, encontrar la posici√≥n y poner texto
for i, genero in enumerate(generos):
    for j, abandono in enumerate(abandono_niveles):
        # Buscar el conteo para ese grupo
        fila = tabla[(tabla['genero'] == genero) & (tabla['abandono'] == abandono)]
        if not fila.empty:
            count = int(fila['count'])
            total = total_dict[genero]
            porcentaje = count / total * 100

            # Calcular posici√≥n de la barra en el gr√°fico:
            # El ancho de cada barra es aproximadamente 0.8 / n√∫mero de hues (a ojo)
            n_hues = len(abandono_niveles)
            bar_width = 0.8 / n_hues
            # Posici√≥n x base del grupo i en el eje x
            x = i - 0.4 + bar_width/2 + j*bar_width

            # Altura de la barra
            altura = count

            ax.text(x, altura + max(tabla['count']) * 0.02, f'{count}\n({porcentaje:.1f}%)',
                    ha='center', va='bottom', fontsize=10)

plt.ylim(0, tabla['count'].max() * 1.15)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Abandono: S√≠
abandono = df_normalizado[df_normalizado['abandono'] == 'S√≠']['tipo_contrato'].value_counts()
axes[0].pie(
    abandono,
    labels=abandono.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("Set2"),
    wedgeprops={'width': 0.4}
)
axes[0].set_title("Clientes que Abandonaron")

# Abandono: No
no_abandono = df_normalizado[df_normalizado['abandono'] == 'No']['tipo_contrato'].value_counts()
axes[1].pie(
    no_abandono,
    labels=no_abandono.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("Set2"),
    wedgeprops={'width': 0.4}
)
axes[1].set_title("Clientes que No Abandonaron")

plt.suptitle("Comparativa de Evasi√≥n por Tipo de Contrato", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Crear tabla de contingencia con conteos
tabla = pd.crosstab(df_normalizado['metodo_pago'], df_normalizado['abandono'])

# Calcular proporciones (opcional para que sea proporcional)
tabla_prop = tabla.div(tabla.sum(axis=1), axis=0)

# Colores para abandono "No" y "S√≠"
colores = ['#66c2a5', '#fc8d62']

fig, ax = plt.subplots(figsize=(10,6))

# Graficar barras apiladas proporcionales
tabla_prop.plot(kind='bar', stacked=True, color=colores, ax=ax)

ax.set_title('Proporci√≥n de Evasi√≥n por M√©todo de Pago', fontsize=16)
ax.set_ylabel('Proporci√≥n de Clientes')
ax.set_xlabel('M√©todo de Pago')
ax.legend(title='Abandono', fontsize=12, title_fontsize=14)
plt.xticks(rotation=45)

# Agregar etiquetas con porcentaje encima de cada secci√≥n
for i, metodo in enumerate(tabla_prop.index):
    bottom = 0
    for abandono in tabla_prop.columns:
        altura = tabla_prop.loc[metodo, abandono]
        if altura > 0.05:  # Mostrar solo si es >5%
            ax.text(i, bottom + altura/2, f'{altura:.0%}', ha='center', va='center', color='white', fontsize=11)
        bottom += altura

plt.tight_layout()
plt.show()
```

## **ü§ñ Modelado Predictivo**

```python
#Separaci√≥n de Datos
from sklearn.model_selection import train_test_split

# Separar variables predictoras y variable objetivo
X = df_codificado.drop(columns=['abandono', 'abandono_label'])  # evitamos incluir la clase y su etiqueta textual
y = df_codificado['abandono']

# Divisi√≥n 70% entrenamiento / 30% prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

# Verificar tama√±os
print(f"üì¶ Tama√±o de entrenamiento: {X_train.shape[0]} muestras")
print(f"üß™ Tama√±o de prueba: {X_test.shape[0]} muestras")

# Verificar distribuci√≥n de clases
print("\nDistribuci√≥n en y_train:")
print(y_train.value_counts(normalize=True))
print("\nDistribuci√≥n en y_test:")
print(y_test.value_counts(normalize=True))
```

```python
#Creaci√≥n de Modelos
#Modelo 1: Regresi√≥n Log√≠stica con Normalizaci√≥n
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score

# Escalado
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Modelo
modelo_log = LogisticRegression(max_iter=1000, random_state=42)
modelo_log.fit(X_train_scaled, y_train)

# Predicci√≥n
y_pred_log = modelo_log.predict(X_test_scaled)
y_prob_log = modelo_log.predict_proba(X_test_scaled)[:, 1]

# Evaluaci√≥n
print("üìä Regresi√≥n Log√≠stica - Evaluaci√≥n")
print(classification_report(y_test, y_pred_log))
print("AUC-ROC:", roc_auc_score(y_test, y_prob_log))
```

```python
#Modelo 2: Random Forest sin Normalizaci√≥n
from sklearn.ensemble import RandomForestClassifier

# Modelo
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_rf.fit(X_train, y_train)

# Predicci√≥n
y_pred_rf = modelo_rf.predict(X_test)
y_prob_rf = modelo_rf.predict_proba(X_test)[:, 1]

# Evaluaci√≥n
print("\nüå≤ Random Forest - Evaluaci√≥n")
print(classification_report(y_test, y_pred_rf))
print("AUC-ROC:", roc_auc_score(y_test, y_prob_rf))
```

```python
#Evaluaci√≥n de los Modelos
#Evaluaci√≥n de Regresi√≥n Log√≠stica
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay

# Regresi√≥n Log√≠stica
print("üìå Regresi√≥n Log√≠stica")

accuracy_log = accuracy_score(y_test, y_pred_log)
precision_log = precision_score(y_test, y_pred_log)
recall_log = recall_score(y_test, y_pred_log)
f1_log = f1_score(y_test, y_pred_log)

print(f"Accuracy: {accuracy_log:.4f}")
print(f"Precision: {precision_log:.4f}")
print(f"Recall: {recall_log:.4f}")
print(f"F1-score: {f1_log:.4f}")

# Matriz de Confusi√≥n
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_log, cmap='Blues')
plt.title("Matriz de Confusi√≥n - Regresi√≥n Log√≠stica")
plt.show()
```

```python
#Evaluaci√≥n de Random Forest
# Random Forest
print("üå≤ Random Forest")

accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)

print(f"Accuracy: {accuracy_rf:.4f}")
print(f"Precision: {precision_rf:.4f}")
print(f"Recall: {recall_rf:.4f}")
print(f"F1-score: {f1_rf:.4f}")

# Matriz de Confusi√≥n
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_rf, cmap='Greens')
plt.title("Matriz de Confusi√≥n - Random Forest")
plt.show()
```

**An√°lisis cr√≠tico**

üîé ¬øCu√°l modelo tuvo el mejor desempe√±o?

**Random Forest** suele tener mejor F1-score y recall, especialmente con datos no lineales y complejos.

**Regresi√≥n Log√≠stica** puede tener precisi√≥n competitiva, pero menor recall si el desbalance no se ajusta bien.


**‚ö†Ô∏è ¬øOverfitting?**
Si accuracy en entrenamiento >> accuracy en prueba, es probable que haya overfitting.

En Random Forest, puedes detectarlo si el recall o precisi√≥n son alt√≠simos en entrenamiento pero bajos en test.

üîß Posibles soluciones:

Reducir profundidad de los √°rboles (max_depth)

Usar menos estimadores

Aplicar validaci√≥n cruzada o regularizaci√≥n


**‚ö†Ô∏è ¬øUnderfitting?**
Si ambos modelos tienen bajo rendimiento en todas las m√©tricas, es posible que est√©n underfiteando (especialmente la regresi√≥n log√≠stica).

Puede deberse a:

Pocas variables relevantes

No capturar interacciones no lineales

Hiperpar√°metros por defecto poco √≥ptimos

üîß Posibles soluciones:

Incluir nuevas variables derivadas

Ajustar hiperpar√°metros

Probar modelos m√°s complejos como XGBoost
